# 使用  CNN 做手寫數字辨識

這個例子將改採用卷積神經網路 (Convolutional Neural Network) 的方法來做手寫辨識 (MNIST)。

```python
#coding=utf-8
from keras.datasets import mnist
from keras.utils import np_utils
from keras.models import Sequential
from keras.layers import Dense, Activation, Convolution2D, MaxPooling2D, Flatten, Conv2D
import random
import matplotlib.pyplot as plt
%matplotlib inline  

import numpy as np
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
# 顯示 data 與 class
plt.rcParams.update({'font.size': 8})
for i in range(9):
    plt.subplot(3,3,i+1)
    plt.imshow(x_train[i], cmap='gray', interpolation='none')
    plt.title("Class {}".format(y_train[i]))
plt.tight_layout()
```

```python
# 資料預處理
x_train = x_train.reshape(-1, 1, 28, 28)/255. # -1 代表 sample 的個數, 1 代表 channel 黑白照片只有 1 個 channel
x_test = x_test.reshape(-1, 1, 28, 28)/255.   #  彩色照片則有 R, G, B 共 3 個 channel
y_train = np_utils.to_categorical(y_train, num_classes=10)
y_test = np_utils.to_categorical(y_test, num_classes=10)
```

```python
x_train.shape 
```

```python
# 開始建構 CNN 模型
model = Sequential()

# 設定第一個卷積層，抽取 32 個特徵 
# Conv layer 1 output shape (32, 28, 28)
model.add(Conv2D(
    filters=32,
    kernel_size=(5,5),
    padding='same',
    input_shape=(1,28,28)
))

# Activation funtion for layer 1
model.add(Activation('relu'))

# Pooling layer 1 (max pooling) output shape (32, 14, 14)
model.add(MaxPooling2D(
    pool_size=(2, 2),
    strides=(2, 2),
    padding='same',
))
```

```python
# Convolution layer 2 output shape (64, 14, 14)
model.add(Conv2D(
    filters=64,
    kernel_size=(5,5),
    padding='same',
))

# Activation funtion for layer 2
model.add(Activation('relu'))

# Pooling layer 2 (max pooling) output shape (64, 7, 7)
model.add(MaxPooling2D(
    pool_size=(2, 2),
    strides=(2, 2),
    padding='same',
))
```

```python
# Fully connected layer 1 input shape (64 * 7 * 7) = (3136), output shape (1024)
# 先使用 Flatten 把三維的層攤平成一維的
model.add(Flatten())

# 設定神經網路連接層架構
model.add(Dense(1024))

# 設定 Activation Function
model.add(Activation('relu'))
```

```python
# Fully connected layer 2 to shape (10) for 10 classes (最後一層)
# 設定神經網路連接層架構
model.add(Dense(10))

# 設定 Activation Function
model.add(Activation('softmax'))
```

```python
## 設定優化器 optimizer 與評估函數 loss function 的種類
model.compile(optimizer='adam',
             loss='categorical_crossentropy',
             metrics=['accuracy'])
```

```python
## 開始訓練：設定 epochs 與 batch_size
# epochs: the number of epochs to train the model
# batch_size: Number of samples per gradient update
model.fit(x_train, y_train, epochs=2, batch_size=32)
```

```python
## 印出在 testing dataset 的 loss 與 accuracy
loss, accuracy = model.evaluate(x_test, y_test)
```

```python
print('test loss: ', loss)
print('test accuracy: ', accuracy)
```

顯示測試資料的結果圖

```python
# The predict_classes function outputs the highest probability class
# according to the trained classifier for each input example.
# 取得所有 test data 的分類結果分別屬於哪一個 Class 
predicted_classes = model.predict_classes(x_test)

# Check which items we got right / wrong
correct_indices = np.nonzero(predicted_classes == y_test.argmax(1))[0]
incorrect_indices = np.nonzero(predicted_classes != y_test.argmax(1))[0]
```

```python
# 印出前十筆分類結果查看
predicted_classes[:10]
```

```python
# 將 np_utils.to_categorical 轉回原形的方法
y_test.argmax(1)[:10] # 印出前十筆資料
```

```python
# 顯示分類正確的圖片
plt.rcParams.update({'font.size': 8}) # 設定文字大小
plt.figure()
for i, correct in enumerate(correct_indices[:9]):
    plt.subplot(3,3,i+1)
    plt.imshow(x_test[correct].reshape(28,28), cmap='gray', interpolation='none')
    plt.title("Predicted {}, Class {}".format(predicted_classes[correct], y_test.argmax(1)[correct],fontsize=8))
plt.tight_layout() # 設定圖片排版
```

```python
# 顯示分類錯誤的圖片
plt.rcParams.update({'font.size': 8}) # 設定文字大小
plt.figure()
for i, incorrect in enumerate(incorrect_indices[:9]):
    plt.subplot(3,3,i+1)
    plt.imshow(x_test[incorrect].reshape(28,28), cmap='gray', interpolation='none')
    plt.title("Predicted {}, Class {}".format(predicted_classes[incorrect], y_test.argmax(1)[incorrect]),fontsize=8)
plt.tight_layout() # 設定圖片排版
```
